# Logistics Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
class Logistics_Regression():
    
    def __init__(self, learning_rate, no_of_iteration):
        self.learning_rate = learning_rate
        self.no_of_iteration = no_of_iteration
    
    #fit function to the model
    def fit(self, X, y):
        self.m, self.n = X.shape
        
        #weight and bias value
        self.w = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.y = y
        
        #gradient Descent
        for i in range(self.no_of_iteration):
            self.update_weight() 
    
    def update_weight(self):
        #Sigmoid Function
        y_hat = 1 / (1 + np.exp(-(self.X.dot(self.w) + self.b)))  # wx + b
        
        #Derivation
        dw = (1/self.m) * np.dot(self.X.T, (y_hat - self.y))
        db = (1/self.m) * np.sum(y_hat - self.y)
        
        #updating weight and bias using Gradient Descent
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db
        
    def predict(self, X):
        y_pred = 1 / (1 + np.exp(-(X.dot(self.w) + self.b)))  # wx + b       
        y_pred = np.where(y_pred > 0.5, 1, 0)
        return y_pred
```

```python
X = data.drop(columns='target',axis=1)
```

```python
y = data['target']
```

```python
se = StandardScaler()
X = se.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,shuffle=True,stratify=y,random_state=7)
```

```python
model = Logistics_Regression(learning_rate=0.01,no_of_iteration=1000)
```

```python
model.fit(X_train,y_train)
```

```python
X_train_pred = model.predict(X_train)
prediction_train = accuracy_score(X_train_pred,y_train)
print("Accuracy score for Train Data : ",prediction_train)
```

```python
X_test_pred = model.predict(X_test)
prediction_test = accuracy_score(X_test_pred,y_test)
print("Accuracy score for Test Data : ",prediction_test)
```
